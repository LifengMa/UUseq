```python
import pandas as pd
import scanpy as sc
import loompy
import os
```

```python
file=os.listdir("/media/ggj/UU-4T-1/E100050281-1/L01/OUTPUT/0812/")
file
```

```python
file=file[1:5]
```

```python
file
```

```python
data_path="/media/ggj/UU-4T-1/E100050281-1/L01/OUTPUT/0812/"
loom_path='/out/'
save_path="/media/ggj/ggjlab2/UUseq/brain/RNA_velocity/loom/"
```

```python
for temp in file:
    loomfile=os.listdir(data_path+temp+loom_path)[0]
    #print(loomfile)
    data=sc.read_loom(data_path+temp+loom_path+loomfile)
    cellID=data.obs
    cellname=[index.split(":")[1][0:30] for index,row in cellID.iterrows()]
    cellname=pd.DataFrame(cellname[0:])
    cellname.columns=['cellID']
    data.obs.index='col'+temp+'_0812.'+cellname['cellID']
    data.write_loom(save_path+'col'+temp+'.loom')
```
